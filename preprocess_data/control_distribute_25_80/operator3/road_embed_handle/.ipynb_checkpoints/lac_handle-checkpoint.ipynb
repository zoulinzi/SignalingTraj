```python
import pandas as pd
import copy

# 读取轨迹数据
df_bs = pd.read_csv('../traj_station_handle/station0.csv')
```

```python
df_bs.head()
```

```python
# 进行one-hot编码
df_bs_one_hot = pd.get_dummies(df_bs['c_lac'], prefix='c_lac')

# 将编码后的数据合并到原数据框中
df_bs = pd.concat([df_bs, df_bs_one_hot], axis=1)
df_bs.head()
```

```python
# 获取实际的基站数量（数据行数）和编码维度（唯一c_lac值的数量）
num_stations = df_bs.shape[0]
encoding_dim = df_bs.shape[1]

# 将DataFrame转换为numpy数组
vector = df_bs_one_hot.to_numpy()
vector.shape
```

```python
import numpy as np
np.save('../vgae/lac_feature.npy', vector)
```
